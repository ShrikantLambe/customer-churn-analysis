# Exploratory Data Analysis (EDA) for Telco Customer Churn
This notebook demonstrates how to perform EDA using a Jupyter Notebook (.ipynb) file. We will load the data, check for missing values, analyze churn distribution, and visualize key relationships.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Churn distribution
sns.countplot(x='Churn', data=df)
plt.title('Churn Distribution')
plt.show()
```

```python
# Churn vs Contract Type
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn by Contract Type')
plt.show()
```

```python
# Churn vs Tenure
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=30)
plt.title('Churn by Tenure')
plt.show()
```

```python
# Churn vs Monthly Charges
sns.kdeplot(data=df, x='MonthlyCharges', hue='Churn', fill=True, common_norm=False, alpha=0.5)
plt.title('Churn by Monthly Charges')
plt.show()
```

```python
# Correlation heatmap for numerical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(8,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Numerical Features)')
plt.show()
```
